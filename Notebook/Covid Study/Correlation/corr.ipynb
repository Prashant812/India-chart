```python
def correlation(dataset1,dataset2):
  df_mobility = pd.read_csv(dataset1)
  df_mobility['sub region'] = df_mobility['sub region'].str.upper()
  df_mobility.rename(columns = {'sub region':'Districts'}, inplace = True)
  df_cases = pd.read_csv(dataset2)
  df_cases['Districts'] = df_cases['Districts'].str.upper()
  new_df = pd.merge(df_mobility, df_cases, on= "Districts")
  new_df.drop(['Unnamed: 0_x', 'Unnamed: 0_y','date'], axis=1,inplace=True)
  
  return new_df.corr()['Mean Cases']
```

```python
for i in dataset_url_mobility:
  for j in dataset_url_cases:
    print(correlation(i,j))
```

```python
def correlation_vacc(dataset1,dataset2):
  df_mobility = pd.read_csv(dataset1)
  df_mobility['sub region'] = df_mobility['sub region'].str.upper()
  df_mobility.rename(columns = {'sub region':'District'}, inplace = True)
  df_vacc = pd.read_csv(dataset2)
  df_vacc.drop(df_vacc.columns.difference(['Mean Vaccination','District']), 1, inplace=True)
  df_vacc['District'] = df_vacc['District'].str.upper()
  new_df = pd.merge(df_mobility, df_vacc, on= "District")
  #new_df.drop(['Unnamed: 0_x', 'Unnamed: 0_y','date'], axis=1,inplace=True)
  
  return new_df.corr()['Mean Vaccination']
```

```python
for i in dataset_url_mobility:
  for j in dataset_url_vaccination:
    print(correlation_vacc(i,j))
```
