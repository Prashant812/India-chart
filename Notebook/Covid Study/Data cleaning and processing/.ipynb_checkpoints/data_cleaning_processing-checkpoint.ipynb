```python
import pandas as pd
```

```python
df1 = pd.read_csv("2021_Mobility.csv")
```

```python
df1.head()
```

```python
district_df1 = df1['sub region']
unique_district_df1 = set(district_df1)
unique_district_df1 = list(unique_district_df1)
len(unique_district_df1)
```

```python
df_jan = pd.read_excel("jan_2021.xlsx")
df_jan.head()
```

```python
df1.iloc[0,:]
```

```python
for i in range(0, len(unique_district_df1)):
    for j in range(0, len(unique_district_df1)):
        if(df1.iloc[5*i,7]==unique_district_df1[j]):
            df_jan.loc[len(df_jan.index)] = df1.iloc[5*i,:]
        
```

```python
df_jan.head()
```

```python
df_q1 = pd.read_excel("q1.xlsx")
df_q1.head()
```

```python
for i in range(0, len(unique_district_df1)):
    for j in range(0, len(unique_district_df1)):
        if(df1.iloc[5*i+1,7]==unique_district_df1[j]):
            df_q1.loc[len(df_q1.index)] = df1.iloc[5*i+1,:]
```

```python
df_q1.head()
```

```python
df_q2 = pd.read_excel("q2.xlsx")
df_q2.head()
```

```python
for i in range(0, len(unique_district_df1)):
    for j in range(0, len(unique_district_df1)):
        if(df1.iloc[5*i+2,7]==unique_district_df1[j]):
            df_q2.loc[len(df_q2.index)] = df1.iloc[5*i+2,:]
```

```python
df_q2.head()
```

```python
df_q3 = pd.read_excel("q3.xlsx")
df_q3.head()
```

```python
for i in range(0, len(unique_district_df1)):
    for j in range(0, len(unique_district_df1)):
        if(df1.iloc[5*i+3,7]==unique_district_df1[j]):
            df_q3.loc[len(df_q3.index)] = df1.iloc[5*i+3,:]
```

```python
df_q3.head()
```

```python
df_jan22 = pd.read_excel("jan_2022.xlsx")
df_jan22.head()
```

```python
for i in range(0, len(unique_district_df1)):
    for j in range(0, len(unique_district_df1)):
        if(df1.iloc[5*i+4,7]==unique_district_df1[j]):
            df_jan22.loc[len(df_jan22.index)] = df1.iloc[5*i+4,:]
```

```python
df_jan22.head()
```

```python
df_jan.to_csv('jan21.csv')
df_q1.to_csv('q1.csv')
df_q2.to_csv('q2.csv')
df_q3.to_csv('q3.csv')
df_jan22.to_csv('jan22.csv')
```
