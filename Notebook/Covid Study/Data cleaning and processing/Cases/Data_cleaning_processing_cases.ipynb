```python
import pandas as pd
```

```python
df1 = pd.read_excel("districts.xlsx")
```

```python
df1.head()
```

```python
df1.shape
```

```python
update_df = df1.drop([x for x in range(0,158784)])
```

```python
update_df
```

```python
district_df1 = update_df['District']
unique_district_df1 = set(district_df1)
unique_district_df1 = list(unique_district_df1)
len(unique_district_df1)
```

```python
gk = update_df.groupby('District')
```

```python
gk.first()
```

```python
gk.get_group('Agar Malwa')
```

```python
df_cases_jan21 = pd.DataFrame()
df_cases_jan21['Districts'] = unique_district_df1
mean_cases_jan21 = []
for i in range(0,len(unique_district_df1)):
    mean_cases_jan21.append(round(gk.get_group(unique_district_df1[i]).iloc[0:31,:].mean(axis=0)['Confirmed']))

df_cases_jan21['Mean Cases'] = mean_cases_jan21 
df_cases_jan21.to_csv('jan21_cases.csv')
```

Quarter 1 : Feb 2021 - April 2021

```python
df_cases_q1 = pd.DataFrame()
df_cases_q1['Districts'] = unique_district_df1
mean_cases_q1 = []
for i in range(0,len(unique_district_df1)):
    mean_cases_q1.append(round(gk.get_group(unique_district_df1[i]).iloc[31:120,:].mean(axis=0)['Confirmed']))

df_cases_q1['Mean Cases'] = mean_cases_q1 
df_cases_q1.to_csv('q1_cases.csv')
```

Quarter 2 : May 2021 - july 2021

```python
df_cases_q2 = pd.DataFrame()
df_cases_q2['Districts'] = unique_district_df1
mean_cases_q2 = []
for i in range(0,len(unique_district_df1)):
    mean_cases_q2.append(round(gk.get_group(unique_district_df1[i]).iloc[120:212,:].mean(axis=0)['Confirmed']))

df_cases_q2['Mean Cases'] = mean_cases_q2 
df_cases_q2.to_csv('q2_cases.csv')
```

Quarter 3 : August 2021 - October 2021

```python
df_cases_q3 = pd.DataFrame()
df_cases_q3['Districts'] = unique_district_df1
mean_cases_q3 = []

for i in range(0,len(unique_district_df1)):
    mean_cases_q3.append(gk.get_group(unique_district_df1[i]).fillna(0).iloc[212:304,:].mean(axis=0)['Confirmed'])

df_cases_q3['Mean Cases'] = mean_cases_q3 
df_cases_q3.to_csv('q3_cases.csv')
```
