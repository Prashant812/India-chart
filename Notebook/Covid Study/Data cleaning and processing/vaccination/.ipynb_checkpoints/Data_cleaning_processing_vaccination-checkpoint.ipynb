```python
import pandas as pd
```

```python
df1 = pd.read_csv("cowin_vaccine_data_districtwise.csv")
```

```python
df1.head()
```

```python
df2 = pd.DataFrame()
```

```python
district = df1['District']
df2['District']=district
len(df2.iloc[:,0])
```

```python
i=6
while(i<2896):
    df2[df1.columns[i]] = df1.iloc[:,i]
    i+=10
```

```python
df2
```

```python
df3 = df2.drop(0)
```

```python
df3 = df3.set_index('District')
```

```python
df3.head()
```

```python
df_vac_jan21 = pd.DataFrame()
df_vac_jan21 
```

```python
for i in range(0,16):
    df_vac_jan21[df3.columns[i]] = df3.iloc[:,i]
```

```python
df_vac_jan21 
```

```python
df_vac_jan21 = df_vac_jan21.apply(pd.to_numeric)
```

```python
av_row = round(df_vac_jan21.mean(axis=1))
av_row
```

```python
df_vac_jan21['Mean Vaccination'] = av_row
```

```python
df_vac_jan21.head()
```

```python
df_vac_jan21.to_csv('jan21_vaccination.csv')
```

Quarter 1 - Feb 2021 - April 2021

```python
df_vac_q1 = pd.DataFrame()

for i in range(16,105):
    df_vac_q1[df3.columns[i]] = df3.iloc[:,i]

df_vac_q1 = df_vac_q1.apply(pd.to_numeric)
av_row_q1 = round(df_vac_q1.mean(axis=1))

df_vac_q1['Mean Vaccination'] = av_row_q1
df_vac_q1
```

```python
df_vac_q1.to_csv('q1_vaccination.csv')
```

Quarter 2 - May 2021 - July 2021

```python
df_vac_q2 = pd.DataFrame()

for i in range(105,197):
    df_vac_q2[df3.columns[i]] = df3.iloc[:,i]

df_vac_q2 = df_vac_q2.apply(pd.to_numeric)
av_row_q2 = round(df_vac_q2.mean(axis=1))

df_vac_q2['Mean Vaccination'] = av_row_q2
df_vac_q2.to_csv('q2_vaccination.csv')
df_vac_q2
```
